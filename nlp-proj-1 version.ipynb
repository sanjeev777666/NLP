### Performing tokenization

```python
words=[]
count=0
tokenizer = TweetTokenizer()
for i in range(len(df1_Train)):
    words.append(tokenizer.tokenize(df1_Train["text"][i]))
    for j in range(len(words[i])):
        count+=1
print("total count of words in a given training dataset",count)
```
